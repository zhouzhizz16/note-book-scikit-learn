```python
#scikit-learn example
from sklearn.linear_model.base import LinearRegression
X = [[0, 0], [1, 1], [2, 2]]           #三个样本，每个样本两个特征
y = [0,1,2]                            #y值
clf = LinearRegression() #还有其他可选参数，参加笔记
clf.fit(X,y)
print clf.coef_      #参数
print clf.intercept_  #截距

print clf.predict([[3,4]])  #预测
```

```python
#稀疏矩阵
import numpy as np
from scipy import sparse

row  = np.array([0, 3, 1, 0])
col  = np.array([0, 3, 1, 2])
data = np.array([4, 5, 7, 9])
x_sparse = sparse.coo_matrix((data, (row, col)), shape=(4, 4))
x_array = x_sparse.toarray()
print x_array
print sparse.issparse(x_sparse)
```

```python
##np.vstack 纵向合并两个array
a = np.array([1, 2, 3])
b = np.array([2, 3, 4])
print np.vstack((a,b))
```
